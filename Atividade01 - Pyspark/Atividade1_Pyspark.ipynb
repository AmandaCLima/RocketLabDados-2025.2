###Exercício 1.2 - Retorna a nacionalidade dos jogadores "The Bests"

```python
# Aplique os filtros para retornar os jogadores the bests
the_best = fut_players.filter(
    (fut_players.dribbling > 90) & (fut_players.shooting > 90)
)[['player_id', 'player_name', 'position', 'dribbling', 'shooting', 'overall']]

# Nationalities é um DataDrame da nacionalidade dos jogadores
nationalities = (fut_players.select('player_id', 'player_name', 'nationality'))

# Faça um join dos dois DataDrames, mantendo todos os jogadores de the_best e obtendo suas nacionalidades (dica: a chave é o id)
the_best_nationality = the_best.join(nationalities, 
                                     on='player_id',
                                     how='left') \
                                     .drop(nationalities.player_name)
                                     

the_best_nationality.display()
     
```

##Exercício 2

Use o método .groupby() para descobrir qual país tem o melhor overall médio. Crie a coluna 'avg_overall'

Seu df country_avg_overall deve conter as seguintes colunas:

- nationality
- overall
- avg_overall

```python
# Dessa forma usando o groupBy é possível calcular a media do overall por nacionalidade
avg_overall_by_country = fut_players.groupBy('nationality').agg(F.round(F.avg('overall'), 2).alias('avg_overall'))

# Utilizando um join, é possível obter todas as country, overall, avg_overall na mesma tabela
country_avg_overall = fut_players.join(avg_overall_by_country,
                                        on='nationality',
                                        how='left')['nationality','overall', 'avg_overall'].orderBy(F.col("avg_overall").desc())

# Retornar a nacionalidade com maior overall médio e o overall médio do brasil
melhor = (
    country_avg_overall
    .orderBy(F.col("avg_overall").desc())
    .limit(1)
    .collect()[0]
)

brasil = (
    country_avg_overall
    .filter(F.col("nationality") == "Brazil")
    .collect()[0]
)

display({
    "Melhor overall médio": f"{melhor['nationality']}: {melhor['avg_overall']:.2f}",
    "Overall médio do Brasil": round(brasil['avg_overall'], 2)
})
```

###Exercício 2.1

Crie um racional que retorne a classificação para o jogador de acordo com as instruções abaixo, então aplique isso para o dataframe fut_players.

Observação: considere os limites dentro do intervalo de classificação. exemplo

-50 contém todos os valores menores que 50 e o valor 50 incluso;

51-60 contém todos os valores entre 51 e 60 com os limites [51,60] inclusos no grupo;

e assim por diante ...

```python

"""
    Através do overall do jogador retorne a classificação conforme a seguir:
    Overall -> classification
    -50     -> "Amador"
    51-60   -> "Ruim"
    61-70   -> "Ok"
    71-80   -> "Bom"
    81-90   -> "Ótimo"
    91+     -> "Lenda"
    
    I: int overall
    O: string
"""
# Dica utilize as clasulas when e otherwise
fut_players = (fut_players \
               .withColumn('classification', F.when(F.col('overall') <= 50, 'Amador')
                                            .when((F.col('overall') > 50) & (F.col('overall') <= 60), 'Ruim')
                                            .when((F.col('overall') > 60) & (F.col('overall') <= 70), 'Ok')
                                            .when((F.col('overall') > 70) & (F.col('overall') <= 80), 'Bom')
                                            .when((F.col('overall') > 80) & (F.col('overall') <= 90), 'Ótimo')
                                            .otherwise('Lenda'))
               )

# Contar quantos jogadores há em cada classificação
fut_players.groupBy("classification").count().orderBy("count", ascending=False).display()
```

##Desafio — Montando o Time dos Sonhos do 🇧🇷

O time é composto por 11 jogadores: 1 goleiro, 4 defesa, 4 meio, 2 ataque.


- Goleiro	GK	
- Defesa	CB, LB, RB, LWB, RWB	
- Meio	CM, CDM, CAM, LM, RM	
- Ataque	ST, CF, LW, RW, LF, RF	
- Outros	(demais posições não classificadas)


```python
# Adicionar coluna com a classificação da função de cada jogador
fut_players = fut_players.withColumn('position_group', F.when(F.col('position') == 'GK', 'Goleiro')
                                                        .when(F.col('position').isin(['CB', 'LB', 'RB', 'LWB', 'RWB']), 'Defesa')
                                                        .when(F.col('position').isin(['CM', 'CDM', 'CAM', 'LM', 'RM']), 'Meio')
                                                        .when(F.col('position').isin(['ST', 'CF', 'LW', 'RW', 'LF', 'RF']), 'Ataque')
                                                        .otherwise('Outros'))


# Cria a janela
windowSpec = Window.partitionBy('position_group').orderBy(F.desc('overall'))

# Adiciona uma coluna com o ranking
ranked = fut_players.withColumn('rank', F.row_number().over(windowSpec))

# Seleciona apenas a quantidade de jogadores necessários por função
best_by_pos = ranked.filter(((F.col('rank') == 1) & (F.col('position_group') == 'Goleiro')) |
                            ((F.col('rank') <= 4) & (F.col('position_group') == 'Defesa')) |
                            ((F.col('rank') <= 4) & (F.col('position_group') == 'Meio') |
                             (F.col('rank') <= 2) & (F.col('position_group') == 'Ataque'))
                            ) \
                            .select('nationality', 'player_name', 'position', 'overall', 'rank', 'position_group')\
                            .orderBy('position_group','rank')
best_by_pos.display()

```

Um detelhe é que nesses dados sobre os jogadores, existem jogadores empatados, então qual seria o critério de desempate?

```python
ranked_ordered = ranked.filter(F.col('position_group') == 'Defesa').select('player_name', 'position_group', 'overall', 'rank').orderBy(F.desc('overall'))
ranked_ordered.limit(6).display()
```
